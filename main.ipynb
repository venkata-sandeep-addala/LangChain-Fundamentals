```python
from dotenv import load_dotenv
from langchain_core.prompts import PromptTemplate
from langchain_groq import ChatGroq
import langchain   
```

```python
load_dotenv()
```

```python
summary_prompt = """
    given the information {information} about a person I want you to create:
    1. A short summary
    2. two interesting facts about them
    """
```

```python
prompt_template = PromptTemplate(input_variables=["information"], template=summary_prompt)
```

```python
model = ChatGroq(model="llama-3.3-70b-versatile")
```

```python
chain = prompt_template | model
```

```python
information = """Koduri Srisaila Sri Rajamouli (/ˈrɑːdʒəmaʊlɪ/ RAH-juh-mow-li; born 10 October 1973) is an Indian filmmaker and screenwriter who works in Telugu cinema. Known for his epic, action drama and fantasy genre films, he is one of the most prolific and the highest grossing Indian directors of all time,[2] as well as the highest-paid director in Indian cinema.[3] Rajamouli is a recipient of various national and international honours including a New York Film Critics Circle award, a Critics' Choice Movie Award, two Saturn Awards, four National Film Awards, and five Filmfare Awards. In 2016, the Government of India honoured him with the Padma Shri, for his contributions in the field of art. In 2023, he was included on the Time's list of the 100 most influential people in the world.[4]"""
response = chain.invoke({
    "information": information
})
```

```python
print(response.content)
```
